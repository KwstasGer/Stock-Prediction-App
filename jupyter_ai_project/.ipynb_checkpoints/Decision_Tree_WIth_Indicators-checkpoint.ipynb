```python
import yfinance as yf
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from ta.trend import MACD  # Importing the MACD class
```

```python
# Download stock data from yfinance
symbol = 'AMZN'
start_date = '2018-11-30'
end_date = '2023-11-30'
data = yf.download(symbol, start=start_date, end=end_date)
```

```python
# Bollinger Bands
window = 20
data['Rolling Mean'] = data['Close'].rolling(window=window).mean()
data['Upper Band'] = data['Rolling Mean'] + 2 * data['Close'].rolling(window=window).std()
data['Lower Band'] = data['Rolling Mean'] - 2 * data['Close'].rolling(window=window).std()
```

```python
# Ichimoku
data['Conversion_Line'] = (data['High'].rolling(window=9).max() + data['Low'].rolling(window=9).min()) / 2
data['Base_Line'] = (data['High'].rolling(window=26).max() + data['Low'].rolling(window=26).min()) / 2
data['Leading_Span_A'] = (data['Conversion_Line'] + data['Base_Line']) / 2
data['Leading_Span_B'] = (data['High'].rolling(window=52).max() + data['Low'].rolling(window=52).min()) / 2
```

```python

# MACD
macd_indicator = MACD(data['Close'])
data['MACD'] = macd_indicator.macd()
data['Signal_Line'] = macd_indicator.macd_signal()
```

```python
# McGinley Dynamic
n = 24
data['MD'] = data['Close'] / data['Close'].rolling(window=n).mean() - 1
```

```python
# Future Closing Price as Prediction Variable
data['Next_Close'] = data['Close'].shift(-1)
```

```python
# Remove rows with NaN values due to shifting and rolling calculations
data = data.dropna()
```

```python
# Select features and prediction variable
features = data[['Close', 'Rolling Mean', 'Upper Band', 'Lower Band', 'Conversion_Line', 'Base_Line', 'Leading_Span_A', 'Leading_Span_B', 'MACD', 'Signal_Line', 'MD']]
y = data['Next_Close']
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, y, test_size=0.2, random_state=42)
```

```python
# Create and train the Decision Tree model
model = DecisionTreeRegressor()
model.fit(X_train, y_train)
```

```python
# Predictions on test data
predictions = model.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, predictions)
print(f'Mean Squared Error: {mse}')
```

```python
# Plotting Bollinger Bands
plt.figure(figsize=(12, 6))
plt.plot(data['Close'], label='Close Price', color='blue')
plt.plot(data['Rolling Mean'], label='Rolling Mean', color='red')
plt.plot(data['Upper Band'], label='Upper Band', color='green')
plt.plot(data['Lower Band'], label='Lower Band', color='brown')
plt.title('Bollinger Bands for AMZN')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# Plotting Ichimoku
plt.figure(figsize=(12, 6))
plt.plot(data['Close'], label='Close Price', color='blue')
plt.plot(data['Conversion_Line'], label='Conversion Line', color='red')
plt.plot(data['Base_Line'], label='Base Line', color='green')
plt.plot(data['Leading_Span_A'], label='Leading Span A', color='orange')
plt.plot(data['Leading_Span_B'], label='Leading Span B', color='purple')
plt.fill_between(data.index, data['Leading_Span_A'], data['Leading_Span_B'], where=data['Leading_Span_A'] >= data['Leading_Span_B'], facecolor='lightgreen', interpolate=True)
plt.fill_between(data.index, data['Leading_Span_A'], data['Leading_Span_B'], where=data['Leading_Span_A'] < data['Leading_Span_B'], facecolor='lightcoral', interpolate=True)
plt.title('Ichimoku Indicator for AMZN')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# Plotting MACD
plt.figure(figsize=(12, 6))
plt.plot(data['MACD'], label='MACD', color='blue')
plt.plot(data['Signal_Line'], label='Signal Line', color='red')
plt.bar(data.index, data['MACD'] - data['Signal_Line'], label='MACD Histogram', color='green')
plt.title('MACD for AMZN')
plt.xlabel('Date')
plt.ylabel('MACD Value')
plt.legend()
plt.show()
```

```python
# Plotting McGinley Dynamic
plt.figure(figsize=(12, 6))
plt.plot(data['Close'], label='Close Price', color='blue')
plt.plot(data['MD'], label='McGinley Dynamic', color='red')
plt.title('McGinley Dynamic for AMZN')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
